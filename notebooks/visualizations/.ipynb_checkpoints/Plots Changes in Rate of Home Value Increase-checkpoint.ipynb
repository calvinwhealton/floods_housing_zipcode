PLotting the change in home values when there is a flood insurance claim

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading in pre-processed data
os.chdir('/Users/calvinwhealton/Documents/GitHub/tdi_capstone/data/processed')

claims = pd.read_csv('ts_claims.csv')
zillow = pd.read_csv('zillow_ann_pct_val.csv')
```

```python
claims.head()
```

```python
zillow.head()
```

```python
df_comb = pd.DataFrame(columns=['zip','yr_flood','yr_before','yr_after','claims'])
```

```python
cols_use = range(1997,2019,1)
```

```python
claims['zips'].values
```

```python
claims.set_index('zips',inplace=True)
```

```python
zillow.set_index('GEOID10',inplace=True)
```

```python
for i in claims.index.astype(int):
    if i in zillow.index:
        for j in cols_use:
            if claims.loc[i,str(j)] !=0:
                #print('hi there2')
                pct_yr_before = zillow.loc[i,str(j-1)]
                pct_yr_after = zillow.loc[i,str(j+1)]
                claims_yr = claims.loc[i,str(j)]
                yr_flood = str(j)
                df_comb = df_comb.append({'zip': i, 'yr_flood': yr_flood,'claims':claims_yr,'pct_before':pct_yr_before,'pct_after':pct_yr_after},ignore_index=True)
    
```

```python
df_comb.drop(columns=['yr_before','yr_after'],inplace=True)
```

```python
df_comb.head()
```

```python
df_comb.dropna(inplace=True)
```

```python
df_comb.shape
```

```python
df_comb.head()
```

```python
# making a histogram
n_bins = 20

# Generate a normal distribution, center at x=0 and y=5
x = df_comb['pct_after']

plt.hist(x, bins = n_bins)
plt.title('Increase in Zillow Housing Value Index\nin Year Following Flood Claim')
plt.xlabel('Percentage')
plt.show()
```

```python
# Generate a normal distribution, center at x=0 and y=5
x = df_comb['pct_after'] - df_comb['pct_before']

plt.hist(x, bins = n_bins, color='green')
plt.title('Change in Percentage Increase in ZHVI\nAfter vs Before Flood Claim')
plt.xlabel('Percentage')
plt.show()
```

```python
import numpy as np

plt.scatter(y=df_comb['pct_after'] - df_comb['pct_before'], x=np.log10(df_comb['claims']),color='#fdae6b20')
plt.title('Flood Insurance Claims vs Change in ZHVI')
plt.xticks(ticks=[2,4,6,8],labels=['100','10,000','1,000,000','100,000,000'])
plt.xlabel('Flood Insurance Claims (USD)')
plt.ylabel('Change in Percentage ZHVI')
```

```python
df_comb['GEOID102'] = df_comb['zip'].values.astype(int)
```

```python
df_comb = df_comb.merge(zips_latlong_gdf,on='GEOID102')
```

```python
df_comb.head()
```

```python
def assign_cols_pct_change(l):
    
    col_l = []
    
    for i in l:
        if i < -25:
            col_l.append('#b2182b')
        elif i < -15:
            col_l.append('#d6604d')
        elif i < -5:
            col_l.append('#f4a582')
        elif i < 0:
            col_l.append('#fddbc700')
        elif i > 25:
            col_l.append('#2166ac')
        elif i > 15:
            col_l.append('#4393c3')
        elif i > 5:
            col_l.append('#92c5de')
        elif i > 0:
            col_l.append('#d1e5f000')
    return col_l
```

```python
df_plot = gpd.GeoDataFrame(df_comb,columns=['geometry','diff_before_after'])
```

```python
os.chdir('/Users/calvinwhealton/Documents/GitHub/tdi_capstone/data/geo_data/US_48_COASTLINE')
import geopandas as gpd
us_48_coast = gpd.read_file('US_COASTLINE.shp')
```

```python
plt.rcParams["figure.figsize"] = (40,35)

plt.rcParams["font.size"] = "50"
base = us_48_coast.plot(color='gainsboro',edgecolor='black')
df_plot.plot(ax=base,color=assign_cols_pct_change(df_plot['diff_before_after'].values),marker='o',markersize=60)
plt.xlabel('Longitude',fontsize=50)
plt.ylabel('Latitude',fontsize=50)
plt.title('Zip Code Housing Price Recovery',fontsize=100)
plt.xlim((-130,-65))
plt.ylim((24,50))
```
